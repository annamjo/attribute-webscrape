```python
wait = WebDriverWait(driver,10)
wait.until(EC.visibility_of_all_elements_located((By.CLASS_NAME,"learnable")))
terms = driver.find_elements_by_class_name("learnable")
wordefs = []
for term in terms:
    word = term.find_element_by_class_name("word").text
    print(word)
    try:
        defin = ('adj',term.find_element_by_css_selector("div.def.adjective").text)
    except:
        try:
            defin = ('noun',term.find_element_by_css_selector("div.def.noun").text)
        except:
            defin = ('verb',term.find_element_by_css_selector("div.def.verb").text)
    print(defin)
    if defin[0]=='adj':
        wordefs.append((word,defin))
```

```python
# find synonyms/antonyms
driver.get('https://www.vocabulary.com/dictionary/')
print(driver.title)
wait5 = WebDriverWait(driver,10)
wait5.until(EC.element_to_be_clickable((By.ID,"search")))
search_box = driver.find_element_by_id("search")
```

```python
#to extract word text from synonyms/antonyms list
import time
def findWords(entries):
    wordtexts = []
    for entry in entries:
        wordtext = entry.find_element_by_class_name("word").text
        worddefin = entry.find_element_by_class_name("definition").text
        wordtexts.append((wordtext,worddefin))
    return wordtexts
    
synonyms = []
antonyms = []
syn_root = 'pos:(adjective OR adverb) +synonym:'
ant_root = 'pos:(adjective OR adverb) +antonym:'
for wd in wordefs:
    search_box.clear()
    search_box.send_keys(syn_root+wd[0],Keys.RETURN)
    time.sleep(1.3)
    syns = driver.find_elements_by_class_name("entry")
    synonyms.append(findWords(syns))

    search_box.clear()
    search_box.send_keys(ant_root+wd[0],Keys.RETURN)
    time.sleep(1.3)
    ants = driver.find_elements_by_class_name("entry")
    antonyms.append(findWords(ants))
    # antonyms = scrollToBottom(ants, len(ants))

print(synonyms)
print(antonyms)
```

```python
#if synonyms/antonyms not in words, add to words
for synonym in synonyms:
    for syn in synonym:
        if syn[0] not in data['Attribute']:
            data['Id'].append(id)
            data['Attribute'].append(syn[0])
            data['Definition'].append(syn[1])
            id+=1
for antonym in antonyms:
    for ant in antonym:
        if ant[0] not in data['Attribute']:
            data['Id'].append(id)
            data['Attribute'].append(ant[0])
            data['Definition'].append(ant[1])
            id+=1
```

```python
def addToCol(words, lst):
    ids = []
    for l in lst:
        ids.append(words.index(l[0])+1)
    
    if len(ids)==0:
        ids = ' '
    return ids

#find ids of synonyms and antonyms from data - include those present in words
for i,word in enumerate(words):
    if i < len(synonyms) and i < len(antonyms): # change later
        data['Synonyms'].append(addToCol(words, synonyms[i]))
        data['Antonyms'].append(addToCol(words, antonyms[i]))    
    else:
        data['Synonyms'].append(' ')
        data['Antonyms'].append(' ')
```
